## Analysis Plot 1
The standard line chart plots the daily value of stock against a specific day of month of the four major defense contractors over the initial six months of Afghanistan war. Daily value of all the afore-mentioned companies is monitored.


```python
#Plotting the High points for each of the Arms selling companies during Afghan War. Back in 2002 GD was the official arms dealer to provide TOMAHAWK missiles to USA.
trace1 = Scatter(
                    x=dfLMT['Date'], y=dfLMT['High'], # Data
                    mode='lines', name='Lockheed Martin Corp' # Additional options
                   )
trace2 = Scatter(x=dfRTN['Date'], y=dfRTN['High'], mode='lines', name='Raytheon' )
trace3 = Scatter(x=dfBA['Date'], y=dfBA['High'], mode='lines', name='Boeing' )
trace4 = Scatter(x=dfGD['Date'], y=dfGD['High'], mode='lines', name='General Dynamics' )
layout = Layout(title='Shares Of Arms Manufactures during Afghan War',
                   plot_bgcolor='rgb(230, 230,230)')

fig = Figure(data=[trace1, trace2, trace3, trace4], layout=layout)

# Plot data in the notebook
iplot(fig)
```

## Results
The plot encapsulates the steep rise in the value of the mentioned organizations due to the utilization of weapons in the Afghanistan war inflating the demand vs supply ratio.

```python
#Normalizing the 'Volume' column of each dataframe in order to plot the BELL CURVE
dfLMT['Volume'] = dfLMT['Volume']/100
dfLMT['Volume'] = np.round(dfLMT['Volume'], 3)
dfGD['Volume'] = dfGD['Volume']/100
dfGD['Volume'] = np.round(dfGD['Volume'], 3)
dfBA['Volume'] = dfBA['Volume']/100
dfBA['Volume'] = np.round(dfBA['Volume'], 3)
dfRTN['Volume'] = dfRTN['Volume']/100
dfRTN['Volume'] = np.round(dfRTN['Volume'], 3)
dfLMT['Volume']
```

## Analysis Plot 2
 Normal distribution of daily volume of stock traded is plotted against congifured deviation. Abscissa (X-axis) is the deviation in the volume while Ordinate (Y-axis) is the probability of stocks lying within a deviation.  


```python
#Normal distribution of the stocks traded during the entire period for the 4 companies
import plotly.figure_factory as ff

x1 = dfLMT['Volume'].values.tolist()
x2 = dfBA['Volume'].values.tolist() 
x3 = dfGD['Volume'].values.tolist()  
x4 = dfRTN['Volume'].values.tolist()

hist_data = [x1, x2, x3, x4]

group_labels = ['LMT', 'BA', 'GD', 'RTN']

colors = ['#3A4750', '#F64E8B', '#A56CC1', '#A6ACEC']

# Create distplot with curve_type traded to 'normal'
fig = ff.create_distplot(hist_data, group_labels, bin_size=.5, curve_type='normal', colors=colors)

# Add title
fig['layout'].update(title='Normal Distribution of Volume of Stocks traded over the Course')

# Plot!
iplot(fig, filename='Distplot with Normal Curve')
```

## Results
- As the first plot describes the rise in value of stocks, high volumes of shares are traded on the New York Stock Exchange. 
- The normal distribution of the mentioned companies shows a right skewed shift. 
- It indicates a trend of buying a stock at a cheaper price to sell later at a projected surge price due to the ongoing war.
