```python
#Import all the libraries
import requests
import json
import os
import csv
import re
import datetime
import time
import glob
import pandas as pd
import ystockquote
from pprint import pprint
from yahoo_finance import Share
from plotly.graph_objs import Scatter, Figure, Layout
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.plotly as py
from plotly.graph_objs import *
import matplotlib.pyplot as plt
import numpy as np
import operator

relativePath = os.getcwd()
```

```python
list_sections = [];                                          #empty list to store different categories by section_name
d={}                                                         #empty dictionary to store section_name n counts
folder='/final/data'
fileList=[]
sectionCountList= {}
for directory, subDirectory, filename in os.walk(relativePath+folder):
    for f in filename:
        a = os.path.join(directory, f)
        fileList.append(a) #appending all 120 json files in fileList

fileList
```

```python
#creating the list of popular sections having most articles during Afghan war
popular_section = ['World', 'World; Front Page; Washington', 'World; Washington', 'Opinion']
#Creating the list of words related to Afghan War
words= ['killing', 'Afghan', 'War', 'Gulf', 'Terrorism', 'Terror', 'death','Al Qaeda']
```

```python
#Code to count the number of articles containing the words individually from the popular section list
list= {}                                                            #Empty Dictionary
docCount= 0;                                                        #For doc count test
fileCount= 0;
for word in words:
    count= 0
    #looping popular_section
    for i in range(0,len(popular_section)):
        for fileName in fileList:
            with open(fileName) as r:                                #opening each files and giving alias as r
                json_data = json.load(r)                             #loading the data of each file into json_data
                json_response=json_data['response']                  #getting response from each page
                json_docs=json_response['docs']                      #getting all articles from every page
                fileCount += 1;
                for doc in json_docs: 
                    docCount += 1
                    if doc['lead_paragraph'] is not None and word in doc['lead_paragraph']:
                        count += 1
    list[word]= count
print(list)
print(str(docCount/len(popular_section)/len(words))+" "+str(fileCount/len(popular_section)/len(words)))
```

```python
#store in CSV file
list
with open(relativePath+"/"+'final/extra/TT2.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in list.items():
        writer.writerow([key, value])
```

## Analyses Plot 1 
The standard plot displays the amount of articles capturing a specific expression.

```python
#Plotting bar graph for words and their artical count
dfWords = pd.read_csv(relativePath+"/"+'final/extra/TT2.csv')
dfWords = dfWords.dropna()
dfWords

trace0 = Bar(
    x=dfWords['Words'],
    y=dfWords['Total_Articles'],
    marker=dict(
        color=['orange', 'rgba(222,45,38,0.8)',
               'blue', 'voilet',
               'green','yellow', 'brown','teal']),
)

data = [trace0]
layout = Layout(
    title='Words in Total Articles',
)

fig = Figure(data=data, layout=layout)
iplot(fig, filename='color-bar')
```

## Inference Plot 1 
- Due to the shocking events of 9/11 and the following military aggression by USA, Afghanistan and war received the maximum article share closely followed by death, killing and Al-Qaeda (the terrorist group responsible for the attacks).


## Analysis Plot 2
- The box plot specifies the highest and lowest values and the mean region for New York Times' individual daily stock price details like high, low, open, close and adjusted close.

```python
#plotting box plot open, close, high, low and adj_close of NYT stocks and shares.
dfWar = pd.read_csv(relativePath+"/"+'final/extra/SharesNYTWarTime.csv')

trace0 = Box(
    y=dfWar['Open'] ,
    name='Open',
    marker=dict(
        color='green',
    ),
    boxmean=True
)

trace1 = Box(
    y=dfWar['Close'] ,
    name='Close',
    marker=dict(
        color='red',
    ),
    boxmean=True
)
trace2 = Box(
    y=dfWar['High'] ,
    name='High',
    marker=dict(
        color='blue',
    ),
    boxmean=True
)
trace3 = Box(
    y=dfWar['Low'] ,
    name='Low',
    marker=dict(
        color='yellow',
    ),
    boxmean=True
)

trace5 = Box(
    y=dfWar['Adj_Close'] ,
    name='Adj_Close',
    marker=dict(
        color='orsnge',
    ),
    boxmean=True
)

layout = Layout(
    title = "NYT Stocks"
)

data = [trace0, trace1, trace2, trace3, trace5]

fig = Figure(data=data,layout=layout)
iplot(fig)
```

## Inference Plot 2
- Incensed by the events and demanding action, citizens and the government wanted to annihilate the responsible terrorists. 
- The inception of Afghan war lead to a greater interest among citizens in finding out the ground reality and activities carried out by the government. 
- The box plot clearly displays no significant loss in all sectors while the tilt of box towards the high segment reflects a trend of increase in demand of information.
