```python
#Importing directory
import os
#Creating directories and subdirectories from where to access emails
relativePath = os.getcwd()
folders=['midterm']
subfolders= ['data']
sfolders=['enron']
if os.path.isdir(relativePath+'/midterm/data/enron') == 0:
    for folder in folders:
        os.mkdir(os.path.join(relativePath,folder))
        for subfolder in subfolders:
            os.mkdir(os.path.join(relativePath,folder,subfolder))
            for sfolder in sfolders:
                os.mkdir(os.path.join(relativePath,folder,subfolder,sfolder))
else:
    print(relativePath)
```

```python
#importing glob library to read files
import glob
#importing parser to parse emails
from email.parser import Parser
Sender=[]
#defining relative path
userPath=relativePath+'\\midterm\\data\\enron\\users\\maildir\\lay-k\\'
#to fetch the sender of all emails sent to the CEO of Enron
for directory, subDirectory, fileNames in os.walk(userPath):
    for file in fileNames:
        fileD = os.path.join(directory, file)
        with open(fileD, "r") as f:
            data = f.read()
            email = Parser().parsestr(data)
            sender= email['From']
            Sender.append(sender)
#print the sender of all emails sent to the CEO of Enron
print(Sender)
```

```python
#Counting the top 20 senders
sendFreq={}
for s in Sender:
    count=sendFreq.get(s,0)
    sendFreq[s] = count + 1
    sortedDict = sorted(sendFreq.items(), key=lambda x: (-x[1], x[0])) # sorted is used to sort dictionary in descending order.
print(sortedDict[:20])
```
