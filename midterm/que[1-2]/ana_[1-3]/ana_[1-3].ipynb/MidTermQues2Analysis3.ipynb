# Purpose
The purpose of this analysis is to find the parts of speech in the json files and their frequency.
API Used: Article Search

```python
#Import all the libraries
import json
import glob
import re
import os
import nltk
```

```python
#use glob to read the json files
relativePath= os.getcwd() # get the current working directory
#print(relativePath)
folder='/midterm/que[1-2]/ana_[1-3]/Articles/jsonData/ArticleSearch'
fileList=[]
for directory, subDirectory, filename in os.walk(relativePath+folder):
    for f in filename:
        a = os.path.join(directory, f)
        fileList.append(a) #appending all 120 json files in fileList
print(fileList)
```

```python
#clean function to clean 
def clean(x):
    punctuation = re.compile('\W+|(\d+)')
    return  re.match(punctuation, x)

#list to store keywords
keywords=[]
#looping to fetch files from the fileList
for file in range(len(fileList)):
    data= json.loads(open(fileList[file]).read()) #reading data from json using load function
    a=(data["response"]["docs"])
    #looping to fetch keywords from each doc
    for b in range(len(a)):
        x=(data["response"]["docs"][b]["keywords"])
        for y in range(len(x)):
            z=(data["response"]["docs"][b]["keywords"][y]["value"])
            z= re.sub('[^A-Za-z]+', ' ', z)
            z=z.lower()
            z=z.split()
            keywords.extend(z)
print(keywords)
       
```

```python
#tagging POS in keywords list
tag = nltk.pos_tag(keywords)
print(tag)
```

```python
#counting the part of speech
from collections import Counter
count=Counter(tag for word,tag in tag)
print(count)
```

```python
#Generate csv to store POS and their count.
import csv



#Writing to the csv file
file= relativePath+"/"+'midterm/que[1-2]/ana_[1-3]/posCount.csv'

with open(file, 'w', newline='') as fp:
    a = csv.writer(fp, delimiter=',')
    a.writerow(["POS", "Count"])
    a.writerows(count)
```
