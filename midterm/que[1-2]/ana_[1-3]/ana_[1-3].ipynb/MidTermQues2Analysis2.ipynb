```python
#Import all the libraries
import json
import glob
import re
import os
```

```python
#use glob to read the json files
relativePath= os.getcwd() # get the current working directory
#print(relativePath)
folder='/midterm/que[1-2]/ana_[1-3]/Articles/jsonData'
fileList=[]
for directory, subDirectory, filename in os.walk(relativePath+folder):
    for f in filename:
        a = os.path.join(directory, f)
        fileList.append(a)
```

```python
keywords=[]
z= []
for file in range(len(fileList)):
    data= json.loads(open(fileList[file]).read()) #reading data from json using load function
    a=(data["response"]["docs"])
    for b in range(len(a)):
        x=(data["response"]["docs"][b]["keywords"])
        for y in range(len(x)):
            z=(data["response"]["docs"][b]["keywords"][y]["value"])
            z= re.sub('[^A-Za-z]+', ' ', z)
            z=z.lower()
            z=z.split()
            keywords.extend(z)
print(keywords)
```

```python
frequency={}
#counting frequency of each word from headlines                       
for word in keywords:
    count = frequency.get(word,0)
    frequency[word] = count + 1
    #Used Lambda function to sort the dictionary
sortedDict = sorted(frequency.items(), key=lambda x: (-x[1], x[0])) # sorted is used to sort dictionary in descending order.
print(sortedDict)
```

```python
i = 1
priorityList = []

# Looping in order to assign priority to each month.
for priority in range(len(sortedDict)):
    obj1 = sortedDict[priority]
    obj1 = obj1 + (i,)
    priorityList.append(obj1)
    i = i+1
print(priorityList)
```
