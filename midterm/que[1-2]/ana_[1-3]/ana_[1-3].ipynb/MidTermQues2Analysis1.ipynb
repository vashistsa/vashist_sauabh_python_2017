# Purpose
The purpose of this analysis is to find the word frequency in headlines and find the trending topic during the period.
API Used: Archive

```python
#Import all the libraries
import json
import glob
import re
import os
```

Loop in the files in directory and store in fileList

```python
#use glob to read the json files
relativePath= os.getcwd() # get the current working directory
#print(relativePath)
folder='/midterm/que[1-2]/ana_[1-3]/Articles/jsonData'
fileList=[]
#append files
for directory, subDirectory, filename in os.walk(relativePath+folder):
    for f in filename:
        a = os.path.join(directory, f)
        fileList.append(a)
        
```

```python
#define clean function
def clean(x):
    punctuation = re.compile('\W+|(\d+)')
    return  re.match(punctuation, x)

#list to store headlines
headlines=[]
#looping to fetch files from the list
for file in range(len(fileList)):
    data= json.loads(open(fileList[file]).read()) #reading data from json using load function
    a=(data["response"]["docs"])
    #looping to fetch headlines from multiple docs
    for b in range(len(a)):
        x=(data["response"]["docs"][b]["headline"]["main"])
        x= re.sub('[^A-Za-z]+', ' ', x)
        x=x.lower()
        x=x.split()
        headlines.extend(x)
print(headlines)
```

```python
frequency={}
#counting frequency of each word from headlines                       
for word in headlines:
    count = frequency.get(word,0)
    frequency[word] = count + 1
    #Used Lambda function to sort the dictionary
sortedDict = sorted(frequency.items(), key=lambda x: (-x[1], x[0])) # sorted is used to sort dictionary in descending order.
print(sortedDict)
```
