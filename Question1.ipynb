Find out how many words and how many unique words in file 'austen-persuasion.txt'.
Calculate the frequency for each words and show the top 10 most frequent words and their frequency.Try NOT to use FreqDist module.
Keep the punctuation and stop words in this question.



```python
import nltk
from nltk.corpus import gutenberg
nltk.corpus.gutenberg.fileids()
```

```python
words = nltk.corpus.gutenberg.words('austen-persuasion.txt')
words = [word.lower() for word in words] #changing words to lower case
len(words) #total words
```

```python
#total unique words
count = {}                  
for word in words:          
    if word in count:
        count[word] += 1 
    else: 
        count[word] =1
print(count)

    
```

```python
#count of unique words
count = {}
unique= 0
for word in words:
    if word in count:
        count[word] += 1 
    else: 
        count[word] =1
        unique += 1
print(unique)
```

```python
#loop to find top 10 frequent words
count = {}
for word in words:
    if word in count:
        count[word] += 1 
    else: 
        count[word] =1
popular_words = sorted(count, key = count.get, reverse = True) # sorting popular words
top_10 = popular_words[:10]
print(top_10)

    
```
