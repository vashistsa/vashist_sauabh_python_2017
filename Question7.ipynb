# Question7
Use Name corpus, calculate the probability of a name being of a male or female base on the last character(s) of name.
Sample steps:
For char ‘a’:
	total_names = sum(names ending with ‘a’)
	male_names = sum(male names ending with ‘a’)
	female_names = sum(female names ending with ‘a’)
	Probability of name being male = 100 * male_names / total_names
2. Do this for all characters. 
Output must come with proper title (e.g. [a] male: 25%, female: 60%)


```python
import nltk
from nltk.corpus import names
```

```python
nltk.corpus.names.fileids()
```

```python
males= nltk.corpus.names.words('male.txt')
males
```

```python
females= nltk.corpus.names.words('female.txt')
len(females)
```

```python
alpha = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

mtotal = len(males)
ftotal = len(females)
munique = 0
funique = 0
#loop to parse male and female ending with 26 characters 
for x in alpha:
    
    for male in males:    
        if male.endswith(x):
            munique += 1 #count to increase males ending with character x
    for female in females:
        if female.endswith(x):                
            funique += 1 #count to increase females ending with character x
            
    mprob=munique/(munique+funique) #male probability formula
    fprob=funique/(munique+funique) #female probability formula
    print(x, 'Male',"%.2f" %round(mprob,2), '% and', 'Female',"%.2f" %round(fprob,2), '%')
```
